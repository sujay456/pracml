```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
x,y=datasets.make_blobs(centers=5,n_samples=500,n_features=2,random_state=3)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
k=5

color=['green','red','blue','yellow','orange']
clusters={}

for kx in range(k):
    center=10*(np.random.random((x.shape[1],)) -1)
    
    points=[]
    
    clusters[kx]={
        'center':center,
        'points':points,
        'color':color[kx]
    }
```

```python
def euc_dist(x1,x2):
    return np.sqrt(((x1-x2)**2).sum())
def assignTocenter():
    for i in range(x.shape[0]):
        dis=[]
        for kx in range(k):
            dis.append(euc_dist(x[i],clusters[kx]['center']))
        
        whichone=np.argmin(dis)
        
        clusters[whichone]['points'].append(x[i]);

def updateCenters():
    
    for kx in range(k):
        
        points=np.array(clusters[kx]['points'])
        
        if points.shape[0]>0:
            new_center=points.mean(axis=0)
            clusters[kx]['center']=new_center
            clusters[kx]['points']=[]
def plotClusters():
    
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
#             plt.show()
        except:
            pass
        
        uk=clusters[kx]['center']
        
        plt.scatter(uk[0],uk[1],c='black',marker='*')    
```

```python
assignTocenter()
plotClusters()
updateCenters()
```
